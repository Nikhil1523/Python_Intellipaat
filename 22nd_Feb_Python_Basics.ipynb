```python
# Print the sum of all prime numbers between 1 and N.
def sum_of_primes(n):
    """
    Calculates the sum of all prime numbers between 1 and n
    using the Sieve of Eratosthenes algorithm.
    """
    if n < 2:
        return 0
    
    is_prime = [True] * (n + 1)
    is_prime[0] = is_prime[1] = False 
    
    p = 2
    while p * p <= n:
        if is_prime[p]:
            for i in range(p * p, n + 1, p):
                is_prime[i] = False
        p += 1
        
    # --- UPDATED: Using a standard loop instead of the sum() function ---
    prime_sum = 0
    for num in range(2, n + 1):
        if is_prime[num]:
            prime_sum += num 
            
    return prime_sum

# --- Test the function ---
N = 5
total = sum_of_primes(N)

print(f"The sum of all prime numbers between 1 and {N} is: {total}")
```

```python
# Count how many palindromic numbers exist between 1 and N.
def count_palindromes(n):
    """
    Counts how many palindromic numbers exist between 1 and n (inclusive).
    """
    total_palindromes = 0
    
    for num in range(1, n + 1):
        # Convert the number to a string
        num_str = str(num)
        
        # Check if the string is equal to its reverse
        # [::-1] is a Python trick that slices the string backwards
        if num_str == num_str[::-1]:
            total_palindromes += 1
                
    return total_palindromes

# --- Test the function ---
N = int(input('Enter the delimiter value : '))
result = count_palindromes(N)

print(f"There are {result} palindromic numbers between 1 and {N}.")
```

```python
# Create a simple calculator using if-elif for addition, subtraction, multiplication, and division.

def calculatorFunction ():
    i_o = str((input('Enter the function that you would like to perform: ')))
    if (i_o.upper() in ('ADDITION', 'SUBTRACTION', 'MULTIPLICATION', 'DIVISION')):
        n = int(input('Enter the first number: '))
        m = int(input('Enter the second number: '))
        if i_o.upper() == 'ADDITION':
            output = n + m 
        elif i_o.upper() == 'SUBTRACTION':
            if (n > m):
                output = n - m
            else:
                output = m - n
        elif i_o.upper() == 'MULTIPLICATION':
            if ( n != 0 or m != 0):
                output = n * m
            else: 
                output = 0
        else:
            if (n != 0 or m != 0):
                if (n > m):
                    output = abs(n / m)
                else:
                    output = abs(m / n)
            else:
                output = 0
        print(f'You have chosen the function: {i_o.upper()} and the output for the function is {output}')

    else: 
        print('The input given is not supported by the calculator, the supported functions are Addition, Subtraction, Multiplication, Division')
    
calculatorFunction()
```

```python
# Grade a student based on marks (A, B, C, D, F grading system).
def GradingSystem (): 
    marks = int(input('Enter the amount of marks obtained by the student: '))
    if (marks >= 35 and marks <= 50): 
        Grade_Point = 'C'
        print (f'The student has been awarded grade {Grade_Point}')
    elif (marks > 50 and marks <= 75):
        Grade_Point = 'B'
        print (f'The student has been awarded grade {Grade_Point}')
    elif (marks > 75 and marks <= 90):
        Grade_Point = 'A' 
        print (f'The student has been awarded grade {Grade_Point}')
    elif (marks > 90 and marks <= 100):
        Grade_Point = 'EX'
        print (f'The student has been awarded grade {Grade_Point}')
    elif (marks < 35 and marks >= 0):
        Grade_Point = 'F'
        print (f'The student has failed the exam and has been awarded grade {Grade_Point}')
    else: 
        print ('The marks attained is either invalid or is beyond the range of grading')

GradingSystem()
```

```python
# Check whether a year is a leap year.
import calendar
def checkLeapYear():
    yearInput = int(input('Enter the year that has to be checked: '))
    if (yearInput != 0 ):
        if calendar.isleap(yearInput): 
            print (f'The year {yearInput} is a leap year')
        else: 
            print (f'The year {yearInput} is not a leap year')
    else: 
        print ('Kindly enter a valid year')

checkLeapYear()
```
