```python
# Conditional Statements 
# if else, if elif else, else, match

age = int(input('Enter your age : '))

if age > 18 : 
    print ('You can watch the movie!')
else :
    print ('You are not allowed to watch the movie!') 
```

```python
score = int(input('Enter your score: ')) 
if score > 90 and score < 100:
    print ('Grade A+')
elif score > 80 and score <= 90:
    print ('Grade B+')
else :
    print ('Grade C') 
```

```python
# Switch Case 
day = int (input ('Enter the day Number : '))
match day : 
    case 1: print ('Monday')
    case 2: print ('Tuesday')
    case 3: print ('Wednesday')
    case 4: print ('Thursday')
    case default : print ('Anything')
```

```python
# Nested if else 
num = int (input ('Enter a number :')) 
if num >= 0 :
    if num % 2 == 0 :
        print (num, 'is Even and Positive')
    else :
        print (num, 'is Odd and Positive')
else : 
    if num % 2 == 0 : 
        print (num, 'is Even and Negative')
    else : 
        print (num, 'is Odd and Negative')
```

```python
# Loops 
# for, while
name = 'Aryan'
# For running 9 times
for i in range (0, 10): 
  print (f'{i}.Cleanliness is next to Godliness for {name}')
# for running 10 times
for i in range (0, 11): 
  print (f'{i}.Cleanliness is next to Godliness for {name}') 
# for loop with increment amount changed 
for i in range (0, 11, 2): 
    print (f'{i}.Cleanliness is next to Godliness for {name}') 
# for running the code 100 times 
for i in range (1, 101):
    print (f'{i}.Cleanliness is next to Godliness for {name}') 
# for making n number of iterartions, will start from 0
for i in range (50):
    print (f'{i}.Cleanliness is next to Godliness for {name}') 
```

```python
#Break Statement 
name = 'Riya' 
for i in range (5): 
    print (f'{i}.The most beautiful girl in the world is {name}')
    if i == 3:
        continue
    elif i == 4:
        break
    else:
        continue
```

```python
# Guessing game 
word = 'Python' 
guess = ''
gc = 1
while guess != word: 
    while gc != 4:
        guess = str(input('Enter Guess: '))
        guess = guess.lower()
        word = word.lower()
        if guess != word :
            if gc <= 1:
                print (f'You guessed wrong so far {gc} time, Guess Again !!')
                gc += 1
            elif gc == 2:
                print (f'You guessed wrong so far {gc} times, Guess Again !!')
                gc +=1
            else:
                print (f'You have already guess {gc} times, no more chances!!')
                gc +=1
        else : 
            print ('You guessed correctly!')
            break
```
