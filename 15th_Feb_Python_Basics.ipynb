```python
#When type casting, a string value which has decimal points in it cannot be directly converted to int, it has to be converted 
#to float first, and then can be converted into int.
example = '2999.90' 
print ('Datatype is : ', type(example))
#example = int(example) #this will give an error as the string has decimal points 
example = float (example)
print ('Datatype is : ', type(example))
example = int (example) 
print ('Datatype is : ', type(example))
ex = ' ' 
ex = bool (ex) 
print ('Datatype is : ', type(ex))
print (ex) #reads as true because it ex contains a value 
ex = 'False' 
ex = bool (ex)
print (ex) #reads true because ex is a boolean variable with some value inside
```

```python
#Arithematic Operators
#Addition -----> +, Subtraction -----> -, Multiplication -----> *, Division -----> / 
#Modulus -----> %, Floor Division -----> //, Exponential -----> ** 

#Addition 
a, b = 10, 5 
print (a + b)
#Subtraction 
print (a - b)
#Multiplication 
print (a * b)
#Division 
print (a / b) 
#Modulus 
print (a % b)
#Floor Division 
print (a // b)
#Exponential 
print (a ** b)
```

```python
#Assignment Operators 
#+= , -=, *=, /=, %=, //=, **=
a, b = 20, 5
a += b 
print (a)
a -= b
print (a)
a *= b 
print (a)
a /= b
print (a)
a %= b
print (a)
a //= b
print (a)
a **= b
print (a)
```

```python
# Comparision Operator 
# >, <, =>, >=, ==, !=
print (5 != 5)
a = 5
b = 10
s = a > b 
print (s)
```

```python
#Logical Operators 
# and, or, not 

a = True 
b = False 
print (a and b)
print (a or b)
print (not (a and b))
```
